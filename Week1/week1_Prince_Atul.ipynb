## 1.1 Data Types & Control Flow

Write a function `classify_number(n)` that:

- Returns `"negative"` if n < 0
- Returns `"zero"` if n == 0
- Returns `"small positive"` if 0 < n <= 10
- Returns `"large positive"` otherwise

````python
def classify_number(n):
    if n < 0:
        return "negative"
    elif n == 0:
        return "zero"
    elif 0 < n <= 10:
        return "small positive"
    else:
        return "large positive"


# Tests
assert classify_number(-5) == "negative"
assert classify_number(0) == "zero"
assert classify_number(7) == "small positive"
assert classify_number(42) == "large positive"

print("1.1 passed ✓")
````

## 1.3 Exception Handling

Write a function `safe_divide(a, b)` that:

- Returns a/b normally.
- Returns None when dividing by zero.
- Raises TypeError when inputs are not numeric.

````python
def safe_divide(a, b):

    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("Inputs must be numeric")

    try:
        return a / b

    except ZeroDivisionError:
        return None


# Tests
assert safe_divide(10, 2) == 5.0
assert safe_divide(5, 0) is None

try:
    safe_divide('x', 2)
    assert False, "Should have raised TypeError"

except TypeError as e:
    assert str(e) == "Inputs must be numeric"

print("1.3 passed ✓")
````

### Explanation

Exception handling improves program reliability.

- `try` block contains code that may raise an exception.
- `except` block handles errors gracefully.
- ZeroDivisionError occurs when division by zero is attempted.
- TypeError occurs when incompatible data types are used.

## 4.3 Eigenvalues & Eigenvectors

Tasks:

1. Compute eigenvalues and eigenvectors.
2. Verify Av = λv.
3. Plot eigenvectors.
4. Explain their geometric interpretation.

````python
A = np.array([
    [4,1],
    [2,3]
],dtype=float)

eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:")
print(eigenvalues)

print("\nEigenvectors:")
print(eigenvectors)

# Verify Av = λv

for i in range(len(eigenvalues)):
    v = eigenvectors[:,i]

    lhs = A @ v

    rhs = eigenvalues[i] * v

    print(f"\nEigenpair {i+1}")

    print("A @ v =")
    print(lhs)

    print("λv =")
    print(rhs)

# Plot

plt.figure(figsize=(6,6))

origin = [0,0]

for i in range(len(eigenvalues)):
    vec = eigenvectors[:,i] * eigenvalues[i]

    plt.quiver(
        *origin,
        vec[0],
        vec[1],
        angles='xy',
        scale_units='xy',
        scale=1,
        label=f'Eigenvector {i+1}'
    )

plt.xlim(-5,5)
plt.ylim(-5,5)

plt.xlabel("x")
plt.ylabel("y")

plt.title("Eigenvectors")

plt.grid()

plt.legend()

plt.show()

print("4.3 passed ✓")
````

### Geometric Explanation

Eigenvectors are special directions that remain unchanged after a matrix transformation. The matrix stretches or compresses these vectors by a factor equal to their corresponding eigenvalues without changing their direction.

## 4.4 SVD & Dimensionality Reduction

Tasks:

1. Generate a random matrix.
2. Compute SVD.
3. Reconstruct the matrix.
4. Compute rank-1 approximation.
5. Explain the connection between SVD and PCA.

````python
np.random.seed(42)

X = np.random.randn(4,3)

U,S,Vt = np.linalg.svd(
    X,
    full_matrices=False
)

# Reconstruction

X_reconstructed = U @ np.diag(S) @ Vt

# Rank-1 approximation

X_approx = (
    S[0] *
    np.outer(
        U[:,0],
        Vt[0,:]
    )
)

print("U shape:",U.shape)
print("S shape:",S.shape)
print("Vt shape:",Vt.shape)

print("\nSingular Values:")
print(S)

print("\nReconstruction Error:")
print(np.linalg.norm(X-X_reconstructed))

print("\nRank-1 Approximation:")
print(X_approx)

assert np.allclose(
    X,
    X_reconstructed,
    atol=1e-10
)

print("4.4 passed ✓")
````

### SVD → PCA Connection

In Singular Value Decomposition:

- U contains left singular vectors.
- S contains singular values.
- Vt contains right singular vectors.

The rows of Vt represent directions of maximum variance and correspond to principal components in PCA. Singular values indicate the importance of each component. PCA uses these components to reduce dimensionality while preserving maximum information.

## 5.3 Error Metrics

Tasks:

1. Calculate MAE
2. Calculate MSE
3. Calculate RMSE
4. Calculate R²
5. Calculate Adjusted R²

````python
y_true = np.array([3.0,5.0,2.5,7.0,4.5,6.0,1.5,8.0])

y_pred = np.array([2.8,5.2,2.1,7.5,4.0,6.3,2.0,7.8])

n,p = len(y_true),2

mae = np.mean(np.abs(y_true-y_pred))

mse = np.mean((y_true-y_pred)**2)

rmse = np.sqrt(mse)

ss_res = np.sum((y_true-y_pred)**2)

ss_tot = np.sum((y_true-y_true.mean())**2)

r2 = 1-(ss_res/ss_tot)

adj_r2 = 1-(1-r2)*(n-1)/(n-p-1)

print("MAE =",round(mae,4))
print("MSE =",round(mse,4))
print("RMSE =",round(rmse,4))
print("R² =",round(r2,4))
print("Adjusted R² =",round(adj_r2,4))
````

## 5.4 Distribution Testing & Stationarity

Tasks:

1. Perform KS Test.
2. Generate non-stationary series.
3. Perform ADF Test.
4. Difference the series and repeat ADF Test.

````python
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)

s1 = np.random.normal(0,1,200)

s2 = np.random.exponential(1,200)

ks_s1 = stats.kstest(
    s1,
    'norm'
)

ks_s2 = stats.kstest(
    s2,
    'norm'
)

print("KS Test Normal Sample")

print(
    ks_s1.statistic,
    ks_s1.pvalue
)

print("\nKS Test Exponential Sample")

print(
    ks_s2.statistic,
    ks_s2.pvalue
)

t = np.arange(200)

ts = 0.05*t + np.random.normal(
    0,
    1,
    200
)

adf_result = adfuller(ts)

print("\nADF p-value before differencing")

print(adf_result[1])

ts_diff = np.diff(ts)

adf_diff = adfuller(ts_diff)

print("\nADF p-value after differencing")

print(adf_diff[1])
````

### Interpretation

- KS Test checks whether a sample follows a specific distribution.
- ADF Test checks stationarity.
- p-value < 0.05 indicates stationarity.

## 5.5 Model Monitoring Concepts

Tasks:

1. Compute Population Stability Index.
2. Simulate drift.
3. Plot distributions.
4. Define concept drift and covariate drift.

````python
def compute_psi(expected,actual,bins=10):

    epsilon = 1e-10

    breakpoints = np.percentile(
        expected,
        np.arange(0,101,100/bins)
    )

    expected_counts,_ = np.histogram(
        expected,
        bins=breakpoints
    )

    actual_counts,_ = np.histogram(
        actual,
        bins=breakpoints
    )

    expected_pct = expected_counts/len(expected)

    actual_pct = actual_counts/len(actual)

    psi = np.sum(
        (actual_pct-expected_pct)*
        np.log(
            (actual_pct+epsilon)/
            (expected_pct+epsilon)
        )
    )

    return psi


np.random.seed(1)

train_dist = np.random.normal(
    50,
    10,
    1000
)

drift_dist = np.random.normal(
    65,
    12,
    1000
)

psi_value = compute_psi(
    train_dist,
    drift_dist
)

print("PSI =",round(psi_value,4))

fig,ax = plt.subplots()

ax.hist(
    train_dist,
    bins=30,
    alpha=0.5,
    label='Training'
)

ax.hist(
    drift_dist,
    bins=30,
    alpha=0.5,
    label='Production'
)

ax.legend()

ax.set_title(
    "Distribution Shift Visualization"
)

plt.show()
````

### Concept Drift vs Covariate Drift

- **Concept Drift:** Relationship between input and output changes.
- **Covariate Drift:** Input distribution changes.

### PSI Thresholds

- PSI < 0.1 → Stable
- PSI 0.1–0.2 → Minor Shift
- PSI > 0.2 → Major Shift

### Retraining Trigger

Retrain the model when PSI exceeds 0.2 or prediction accuracy drops significantly.

## 6.3 Bayes' Theorem

Spam Filter Example

Given:

- P(Spam)=0.30
- P("free"|Spam)=0.80
- P("free"|Not Spam)=0.05

Tasks:

1. Compute P(Spam|"free")
2. Implement Naive Bayes prediction
3. Explain Prior, Likelihood, Evidence and Posterior

````python
p_spam = 0.30

p_free_given_spam = 0.80

p_free_given_ham = 0.05

p_ham = 1 - p_spam

# Total probability

p_free = (
    p_free_given_spam * p_spam
    +
    p_free_given_ham * p_ham
)

# Bayes theorem

p_spam_given_free = (
    p_free_given_spam * p_spam
) / p_free

print("P(free) =", round(p_free,4))

print("P(Spam|free) =", round(p_spam_given_free,4))


def naive_bayes_predict(
        prior_spam,
        p_word_given_spam,
        p_word_given_ham):

    p_ham = 1 - prior_spam

    evidence = (
        p_word_given_spam * prior_spam
        +
        p_word_given_ham * p_ham
    )

    posterior = (
        p_word_given_spam * prior_spam
    ) / evidence

    return posterior


pred = naive_bayes_predict(
    0.30,
    0.80,
    0.05
)

print("Prediction =", round(pred,4))

assert abs(pred-p_spam_given_free) < 1e-9

print("6.3 passed ✓")
````

### Bayes Term Mapping

| Term | Meaning |
|--------|---------|
| Prior | Initial probability of spam |
| Likelihood | Probability of word "free" given spam |
| Evidence | Overall probability of word "free" |
| Posterior | Probability email is spam after observing "free" |

## 6.4 Central Limit Theorem

Tasks:

1. Generate exponential population.
2. Draw 5000 samples.
3. Compute sample means.
4. Plot histogram.
5. Perform KS test.

````python
np.random.seed(7)

population = np.random.exponential(
    scale=1.0,
    size=100000
)

n_samples = 5000

sample_size = 30

sample_means = np.array([
    np.mean(
        np.random.choice(
            population,
            sample_size
        )
    )
    for _ in range(n_samples)
])

pop_mean = population.mean()

pop_std = population.std()

clt_std = pop_std / np.sqrt(sample_size)

print("Population Mean =", round(pop_mean,4))

print("Population Std =", round(pop_std,4))

print("Sample Mean =", round(sample_means.mean(),4))

print("Sample Std =", round(sample_means.std(),4))

print("CLT Predicted Std =", round(clt_std,4))


fig, ax = plt.subplots()

ax.hist(
    sample_means,
    bins=30,
    density=True,
    alpha=0.6
)

x = np.linspace(
    sample_means.min(),
    sample_means.max(),
    300
)

ax.plot(
    x,
    stats.norm.pdf(
        x,
        pop_mean,
        clt_std
    ),
    color='red'
)

ax.set_title(
    "CLT: Distribution of Sample Means"
)

plt.show()


ks_result = stats.kstest(
    sample_means,
    'norm',
    args=(pop_mean,clt_std)
)

print(
    "KS p-value =",
    round(ks_result.pvalue,4)
)

print(
    "Approximately Normal?",
    ks_result.pvalue > 0.05
)
````

### CLT Reflection

The Central Limit Theorem states that the distribution of sample means approaches a normal distribution regardless of the original population distribution when the sample size is sufficiently large.

This theorem forms the basis of many statistical tests and machine learning algorithms that assume normality.
