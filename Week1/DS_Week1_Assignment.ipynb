```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import norm, binom, poisson

sns.set_style("whitegrid")
```

# Section 4.1: Single Vector Representation

This graph represents a vector originating from the origin and moving upward diagonally.

```python
plt.figure(figsize=(6,6))

plt.arrow(0,0,5,5,
          head_width=0.2,
          head_length=0.3,
          color='blue')

plt.xlim(0,6)
plt.ylim(0,6)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Single Vector")

plt.grid()
plt.show()
```

# Section 4.3: Two Vectors from Common Origin

```python
plt.figure(figsize=(6,6))

plt.arrow(0,0,5,3,
          head_width=0.2,
          head_length=0.3,
          color='red')

plt.arrow(0,0,3,5,
          head_width=0.2,
          head_length=0.3,
          color='green')

plt.xlim(0,6)
plt.ylim(0,6)

plt.title("Two Vectors")
plt.grid()
plt.show()
```

```python
np.random.seed(42)

salary1 = np.random.normal(30000,3000,300)
salary2 = np.random.normal(80000,4000,300)

salary = np.concatenate([salary1,salary2])

plt.figure(figsize=(10,6))

sns.histplot(salary,
             bins=30,
             kde=True,
             color='skyblue')

plt.xlabel("Salary")
plt.ylabel("Frequency")

plt.title("Histogram with Density Curve")

plt.show()
```

```python
months = ["Jan","Feb","Mar","Apr","May","Jun"]
sales = [100,130,170,200,240,300]

plt.figure(figsize=(8,5))

plt.fill_between(months,
                 sales,
                 color='lightgreen',
                 alpha=0.6)

plt.plot(months,sales,
         marker='o')

plt.xlabel("Months")
plt.ylabel("Sales")

plt.title("Area Chart")

plt.show()
```

```python
x = np.linspace(-4,4,1000)
y = norm.pdf(x)

plt.figure(figsize=(8,5))

plt.plot(x,y)

plt.xlabel("Values")
plt.ylabel("Density")

plt.title("Normal Distribution")

plt.show()
```

```python
n = 10
p = 0.5

x = np.arange(0,n+1)
prob = binom.pmf(x,n,p)

plt.figure(figsize=(8,5))

plt.bar(x,prob,color='orange')

plt.xlabel("Successes")
plt.ylabel("Probability")

plt.title("Binomial Distribution")

plt.show()
```

```python
lam = 4

x = np.arange(0,15)
prob = poisson.pmf(x,lam)

plt.figure(figsize=(8,5))

plt.bar(x,prob,color='purple')

plt.xlabel("Occurrences")
plt.ylabel("Probability")

plt.title("Poisson Distribution")

plt.show()
```

```python
sample_means = []

for i in range(1000):
    sample = np.random.exponential(scale=2,size=50)
    sample_means.append(np.mean(sample))

plt.figure(figsize=(10,6))

sns.histplot(sample_means,
             bins=30,
             kde=True)

plt.xlabel("Sample Means")
plt.ylabel("Frequency")

plt.title("Central Limit Theorem")

plt.show()
```

## Central Limit Theorem

The Central Limit Theorem states that regardless of the population distribution, the distribution of sample means tends to become approximately normal as the sample size increases.

It forms the basis of hypothesis testing and statistical inference.
